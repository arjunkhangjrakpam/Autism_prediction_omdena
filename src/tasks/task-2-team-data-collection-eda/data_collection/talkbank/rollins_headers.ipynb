```python
import pylangacq as pla
import pandas as pd
import os

pd.options.display.max_columns = None

import warnings
warnings.filterwarnings("ignore")
```

```python
url = '../../../data/talkbank/Rollins.zip'
rollins = pla.read_chat(url)
rollins.n_files() #how many chat files are present for this data
```

```python
utterrances = rollins.utterances(by_files=True)
```

```python
column_names = ['participant', 'sentence']
df = pd.DataFrame(columns=column_names)

for chat_file in utterrances:
    for utter in chat_file:
        participant = utter.participant
        participant_line = utter.tiers[participant]
        df = df.append({'participant':participant, 'sentence':participant_line}, ignore_index=True)
```

```python
df2['months'] = (df2['Participants.CHI.age'].str.split('.').str[0])
del(df2['Participants.CHI.age'])
df2[['years', 'months']] = df2['months'].str.split(';', expand=True).iloc[:, :2]
df2['years'] = df2['years'].astype(int)
df2['months'] = pd.to_numeric(df2['months'], errors='coerce').fillna(0).astype(int)
df2['age_months'] = 12 * df2['years'] + df2['months']
del(df2['years'])
del(df2['months'])
```

```python
df2.head(1)
```

```python
df2.rename(columns=lambda x: x.replace("Participants.CHI.", ""), inplace=True)
df2.rename(columns={
    'Types': 'types', 'custom': 'name', 'sex': 'gender'}, inplace=True)
df2.head(2)
```

```python
df2['group'] = 'ASD'
```

```python
def merge_columns(row):
    values = [str(row[col]) for col in df2.columns if col.endswith('.name')]
    values = [value for value in values if value != 'nan']
    return ', '.join(values)

df2['investigator'] = df2.apply(merge_columns, axis=1)
df2.drop(columns=[col for col in df2.columns if col.endswith('.name')], inplace=True)
```

```python
rol_mlu = rollins.mlu()
```

```python
file_path = '../../../data/talkbank/chat_files/rollins/'
files = []
for filename in os.listdir(file_path):
    if os.path.isfile(os.path.join(file_path, filename)):
        files.append(os.path.join(file_path, filename))
```

```python
df3 = pd.DataFrame(columns=['filename','all_chat'])
df4 = pd.DataFrame(columns=['filename','all_chat'])

for file in files:
    all = [] #all is used to store the concatenated speech data
    chi = []
    df = pd.read_csv(file) #reading the csv file and storing it in a dataframe
    # if df[df['participant'] == 'CHI']:
    all = ' '.join(df['sentence'].astype(str)) # concatenate the entire speech column into one single string
    temp_df = pd.DataFrame({'filename': [file.split('.')[6][35:]],'all_chat' : [all]}) # storing the result in another dataframe temporarily
    chi = ' '.join(df[df['participant'] == 'CHI']['sentence'].astype(str)) # concatenate the entire speech column into one single string
    temp_chi = pd.DataFrame({'filename': [file.split('.')[6][35:]],'child' : [chi]}) # storing the result in another dataframe temporarily
    df3 = pd.concat([df3, temp_df]) # concatenating the temporary dataframe with our initialised result dataframe
    df4 = pd.concat([df4, temp_chi])
    df3['child'] = df4['child']
df3.index = range(len(df3))
```

```python
df2[['filename', 'all_chat', 'child_chat']] = df3[['filename', 'all_chat', 'child']]
```
