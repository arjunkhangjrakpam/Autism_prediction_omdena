```python
saudi_df_updated = saudi_df[
    ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10',
     'Age_Mons', 'Gender', 'Family_mem_with_ASD',
     'Qchat_10_Score', 'Who_completed_the_test', 'ASD_traits']]
saudi_df_updated['Ethnicity'] = 'Middle Eastern'
saudi_df_updated['src'] = 'Saudi'
saudi_df_updated.head(1)
```

```python
thabtah_df_updated = thabtah_df[
    ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10',
       'Age_Mons', 'Qchat_10_Score', 'Gender', 'Ethnicity',
       'Family_mem_with_ASD', 'Who_completed_the_test', 'ASD_traits']]
thabtah_df_updated['src'] = 'Thabtah'
thabtah_df_updated.head(1)
```

```python
github_df_updated = github_df[
    ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10',
     'Age_Mons', 'Gender', 'Family_mem_with_ASD',
     'Qchat_10_Score', 'Who_completed_the_test', 'ASD_traits', 'Ethnicity']]
github_df_updated['src'] = 'GitHub'
github_df_updated.head(1)
```

```python
def unique(list1):

    # initialize a null list
    unique_list = []

    # traverse for all elements
    for x in list1:
        # check if exists in unique_list or not
        if x not in unique_list:
            unique_list.append(x)
    # print list
    print(sorted(unique_list))
```

```python
unique(saudi_df_updated['Age_Mons']), unique(uci_child_df_updated['Age_Mons']),
unique(thabtah_df_updated['Age_Mons']), unique(github_df_updated['Age_Mons'])
```

```python
merged_qchat_features = pd.concat(
    [saudi_df_updated, uci_child_df_updated,
     thabtah_df_updated,github_df_updated], sort=False)
merged_qchat_features.reset_index(drop=True, inplace=True)
merged_qchat_features
```
