#Github Data

```python
github_df = pd.read_csv(os.path.join(data_folder, 'data_csv.csv'))
```

```python
github_df_child = github_df[github_df.Age_Years < 13]
```

```python
github_df_child.info()
```

```python
#saving the child data to check for all 0 (from a1 to a9) errors
github_df_child.to_csv('drive/My Drive/Omdena/SriLanka/Autism/github_child.csv', index = False)
```

After checking the github data for children, it seems the last ~500 rows have 0 values for a1 to a9 even though some of those datapoints are considered ASD positive(class = Yes). So, those rows will be deleted and then merged with other data. As such out of 1206 datapoints, the first 714 datapoint (that are found to be not-erroneous) are taken

```python
github_df_child_ = github_df_child[:714].copy()
```

```python
github_df_child_updated = github_df_child_[['A1', 'A2','A3','A4','A5','A6','A7','A8','A9','A10_Autism_Spectrum_Quotient',"ASD_traits"]]
github_df_child_updated['Class'] = github_df_child_updated['ASD_traits'].apply(lambda x: 1 if x == 'Yes' else 0)
github_df_child_updated.drop(columns = ['ASD_traits'], inplace = True)
```

```python
github_df_child_updated.rename(columns={ "A10_Autism_Spectrum_Quotient": "A10"}, inplace = True)
github_df_child_updated['src'] = 'github'
```

```python
github_df_child_updated.head()
```

#Combine

```python
#combine all datasets
final_df = pd.concat([saudi_df_updated,uci_child_df_updated,thabtah_df_updated, github_df_child_updated], axis=0)
```
