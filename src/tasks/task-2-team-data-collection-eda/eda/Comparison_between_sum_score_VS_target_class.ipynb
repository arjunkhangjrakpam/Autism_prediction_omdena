# Calculation of  sum score

```python
#calculate the sum score
ad['scores']=(ad.A1_Score	+ad.A2_Score+ad.A3_Score+ad.A4_Score+ad.A5_Score+ad.A6_Score+ad.A7_Score+ad.A8_Score+ad.A9_Score+ad.A10_Score)
ado['scores']=(ado.A1_Score	+ado.A2_Score+ado.A3_Score+ado.A4_Score+ado.A5_Score+ado.A6_Score+ado.A7_Score+ado.A8_Score+ado.A9_Score+ado.A10_Score)
```

```python
td['scores']=(td.A10 +td.A1+td.A2 + td.A3+td.A4+ td.A5+ td.A6+ td.A7+ td.A8+ td.A9)
```

# Q-CHAT-10 TEST

photo_2023-02-27_12-34-18.jpg

```python
# according to the Q_CHAT10 test definde the score
td['sum_score'] =td['scores'].apply(lambda x : 1 if x >3 else 0)
   
```

# AQ-10 TEST

```python
#according to the AQ-10 TEST definde the score 
ad['sum_score'] =ad['scores'].apply(lambda x : 1 if x >6 else 0)
ado['sum_score'] =ado['scores'].apply(lambda x : 1 if x >6 else 0)
```

#Visualization

```python
#merged_saudi_github_thabtah_toddler_child dataset (sum_score VS  class)
import matplotlib.pyplot as plt
fig,ax= plt.subplots(1,2)
sns.countplot(data=td,x='sum_score',hue='src',ax=ax[0])
sns.countplot(data=td,x='Class',hue='src',ax=ax[1])
```

```python
#Autism-Adult dataset (sum_score VS  class)
fig,ax= plt.subplots(1,2)
sns.countplot(data=ad,x='sum_score',ax=ax[0])
sns.countplot(data=ad,x='Class/ASD',ax=ax[1])
```

```python
#Autism-Adolescentdataset (sum_score VS  class)
fig,ax= plt.subplots(1,2)
sns.countplot(data=ado,x='sum_score',ax=ax[0])
sns.countplot(data=ado,x='Class/ASD',ax=ax[1])
```
