# Introduction
### ASD is an inborn trait that causes remarkable difference in the brain. People with ASD often have problems comunicating openly with others, this is specially true for enviroments that feel unsafe to them. Also ASD is associated with biomechanical defficiency, difficulty at learning and even propency to other diseases.

### ASD often is hard to detect at early stages (ages below 13 years).This Exploratory Data Analysis (EDA) aims to find and explain certain characteristics in children that may be linked to possible real life cases of ASD and effectively predict cases unknown by their parents or medical professionals.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
palette = sns.color_palette( palette = 'viridis')
sns.set(rc={'figure.figsize':(11,8)})
```

```python
temp_group=df.groupby(by='Class')['Class'].count()
```

```python
ax=sns.barplot(data=temp_group,x=temp_group.index,y=temp_group, 
            palette=sns.color_palette('viridis',temp_group.unique().shape[0]))
for container in ax.containers:
    ax.bar_label(container)
    
title_font = {'size':'20'}
plt.title("ASD count",fontdict=title_font)
plt.grid(False)
label_font = {'size':'15'}  # Adjust to fit
plt.xlabel("Class", fontdict=label_font);
plt.ylabel('Count', fontdict=label_font);

plt.xticks([0, 1], ['Non ASD', "ASD"])
plt.tick_params(axis='both', which='major', labelsize=10)
plt.show()
```

```python
ax=pd.crosstab(df['src'],df['Class']).plot.bar(stacked=True)

for container in ax.containers:
    ax.bar_label(container)
    
title_font = {'size':'20'}
plt.title("ASD count",fontdict=title_font)
plt.grid(False)
label_font = {'size':'15'}  # Adjust to fit
plt.xlabel("SRC", fontdict=label_font);
plt.ylabel('Count', fontdict=label_font);

plt.tick_params(axis='both', which='major', labelsize=10)
plt.show()
```

```python
title_font = {'size':'20'}
plt.title('Heatmap ASD dataset')
sns.heatmap(df.corr(),vmin=-1,vmax=1,cmap='viridis',annot=True, fmt=".2f")
```

```python
numerical_col = ["A1", "A2", "A3", "A4", "A5", "A6", "A7", "A8", "A9", "A10"]
```

```python
plt.subplots (3, 4, figsize = (15, 15))

plot_num = 1

numerical_df = df.loc[:, numerical_col]

for col in numerical_col: 
    plt.subplot(3, 4, plot_num)
    sns.countplot(data=df, x=col, hue='src')
    plt.title(f'ASD Traits by {col}')  
    plot_num +=1 
plt.tight_layout()
plt.show()
```
