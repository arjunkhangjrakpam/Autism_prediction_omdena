# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from termcolor import colored
```

```python
df = pd.read_csv("../../../data/data_csv.csv")
```

```python
df.head()
```

```python
print(colored("Shape: ","green", attrs=["bold"]), df.shape)
print(colored("Information is:","green", attrs=["bold"]))
print(df.info())
```

```python
df.columns
```

```python
print(colored("Total number of values with respect to each features", "green", attrs=["bold"]), "\n", df.count())
```

```python
print(colored("Null values in each column \n ", "green", attrs=["bold"]), df.isnull().sum())
```

```python
df.dtypes
```

```python
cat_df = df[['Speech Delay/Language Disorder','Learning disorder', 'Genetic_Disorders', 'Depression',
'Global developmental delay/intellectual disability',
'Social/Behavioural Issues','Anxiety_disorder', 'Sex', 'Ethnicity', 'Jaundice',
'Family_mem_with_ASD', 'Who_completed_the_test', 'ASD_traits']]

num_df = df[['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8',
       'A9', 'A10_Autism_Spectrum_Quotient', 'Social_Responsiveness_Scale',
       'Age_Years', 'Qchat_10_Score', 'Childhood Autism Rating Scale']]
```

```python
for i in cat_df.columns:
    print(colored(i,"green", attrs=["bold"]), df[i].unique())
```

# ASD Traits

```python
ax = sns.barplot(x=df["ASD_traits"].value_counts().index, y=df["ASD_traits"].value_counts())
ax.bar_label(ax.containers[0])
plt.show()
```

## ASD Traits & Categorical Features

```python
for i in cat_df.columns:
    ax = sns.barplot(x=cat_df[i].value_counts().index, y=cat_df[i].value_counts())
    plt.title(i)
    plt.xticks(rotation=90)
    ax.bar_label(ax.containers[0])
    plt.show()
```

## Ethnicity and ASD Traits

```python
pivot_ethnicity = pd.pivot_table(df, index="Ethnicity", columns="ASD_traits", values="Qchat_10_Score")
pivot_ethnicity
```

```python
ax = pivot_ethnicity.plot(kind="bar")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])

plt.show()
```

```python
test = pd.pivot_table(df, index='Who_completed_the_test', columns="ASD_traits", values="Qchat_10_Score")
test
```

```python
ax = test.plot(kind="bar")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.show();
```

## ASD Traits & Numerical Features

```python
num_df = df[['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8',
       'A9', 'A10_Autism_Spectrum_Quotient', 'Social_Responsiveness_Scale',
       'Age_Years', 'Qchat_10_Score', 'Childhood Autism Rating Scale']]
```

```python
for i in num_df.columns:
    ax = sns.histplot(num_df[i])
    plt.title(i)
    ax.bar_label(ax.containers[1])
    plt.show()
```

```python
sns.heatmap(num_df.corr())
```

```python
pivot_num = pd.pivot_table(df, index="ASD_traits", values=['Social_Responsiveness_Scale',
       'Age_Years', 'Qchat_10_Score', 'Childhood Autism Rating Scale'])
pivot_num
```

```python
ax = pivot_num.plot(kind="bar")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
ax.bar_label(ax.containers[2])
ax.bar_label(ax.containers[3])
ax.bar_label(ax.containers[0])
plt.show();
```
